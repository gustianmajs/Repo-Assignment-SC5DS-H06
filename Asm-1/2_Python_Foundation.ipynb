# Question 2 and 3

2. Hitunglah BMI setiap siswa dengan rumus di soal
3. Buatlah program untuk menghitung BMI!

```python
# mengonversi kolom 'tinggi_badan' dari cm ke meter
df_H['tinggi_badan_m'] = df_H['tinggi_badan'] / 100
```

Karena kolom tinggi badan masih dalam bentuk cm sementara dalam soal menggunakan m maka di sini data di kenversi terlebih dahulu menjadi m dengan nama tabel tinggi_badan_m.

```python
# fungsi untuk menghitung BMI
def hitung_bmi(berat_badan, tinggi_badan):
    return round(berat_badan / (tinggi_badan ** 2), 2)
```

```python
# Memindahkan kolom 'tinggi_badan_m' ke sebelah kolom 'tinggi_badan'
df_H.insert(df_H.columns.get_loc('tinggi_badan') + 1, 'tinggi_badan_m', df_H.pop('tinggi_badan_m'))
```

# Question 4

4. Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.

```python
# menghitung BMI untuk setiap baris dalam DataFrame
df_H['bmi'] = df_H.apply(lambda row: hitung_bmi(row['berat_badan'], row['tinggi_badan_m']), axis=1)
```

```python
df_H
```

Terlihat pada hasil nomor 4 kolom bmi dan outputnya sudah sesuai yang di inginkan.

# Question 5

Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

- Normal: 18.5-24.9
- Overweight: 25.0-29.9
- Obesity: 30.0-34.9
- Underweight: < 18.5


```python
# menambahkan kolom 'status_bmi' berdasarkan kategori BMI
def status_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25.0:
        return 'Normal'
    elif 25.0 <= bmi < 30.0:
        return 'Overweight'
    elif 30.0 <= bmi <= 34.9:
        return 'Obesity'
    else:
        return 'Obesity'
```

```python
df_H['status_bmi'] = df_H['bmi'].apply(status_bmi)
```

```python
df_H
```

Terlihat dari data tersebut bahwa status_bmi setiap orang dari dataframe.
